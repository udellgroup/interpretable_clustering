```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('rep_nodes_data.csv')
df = df[df['time'] < 900]
df['method'] = [0]*len(df)
del df['Unnamed: 0']
```

```python
df
```

```python
plt.scatter(df['num_nodes'], df['time']) 
plt.xlabel("Number of Nodes")
plt.ylabel("Time")
```

```python
x = df.sort_values(by=['num_nodes'])
plt.plot(x['num_nodes'], x['time']) 
plt.xlabel("Number of Nodes")
plt.ylabel("Time")
```

```python
df.set_index('num_nodes', inplace=True)
df.groupby('num_rows')['time'].plot(legend=True)
```
